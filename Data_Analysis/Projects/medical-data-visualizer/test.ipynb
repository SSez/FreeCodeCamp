```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('medical_examination.csv')
```

```python
overweight = df['overweight'] = (df['weight'] / ((df['height'] / 100) ** 2 )).apply(lambda x: 1 if x > 25 else 0)
```

```python
overweight
```

```python
# Normalize data by making 0 always good and 1 always bad. If the value of 'cholesterol' or 'gluc' is 1, make the value 0. If the value is more than 1, make the value 1.
df['cholesterol'] = df['cholesterol'].apply(lambda x: 0 if x == 1 else 1)
```

```python
df['gluc'] = df['gluc'].apply(lambda x: 0 if x == 1 else 1)
```

```python
df['cholesterol']
```

```python
df['gluc']
```

```python
df_cat = pd.melt(df, id_vars = ['cardio'], value_vars = ['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])
```

```python
df_cat
```

```python
df_cat = df_cat.groupby(['cardio', 'variable', 'value'], as_index = False).size().rename(columns={'size' : 'total'})
```

```python
df_cat
```

```python
fig = sns.catplot(data = df_cat, x = 'variable', y = 'total', hue = 'value', kind = 'bar', col = 'cardio').fig
```

```python
df_heat = df[
  (df['ap_lo'] <= df['ap_hi']) &
  (df['height'] >= df['height'].quantile(0.025)) &
  (df['height'] <= df['height'].quantile(0.975)) &
  (df['weight'] >= df['weight'].quantile(0.025)) &
  (df['weight'] <= df['weight'].quantile(0.975))
]
```

```python
df_heat
```

```python
corr = df_heat.corr(method='pearson')
```

```python
corr
```

```python
mask = np.triu(corr)
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
```

```python
sns.heatmap(corr, linewidths = 1, center = 0, annot = True, square = True, mask = mask, fmt = '0.1f', cbar_kws = {'shrink' : 0.5})
```
